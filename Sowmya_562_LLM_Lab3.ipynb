```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load the pre-trained GPT-2 model and tokenizer
model_name = 'gpt2-medium'
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Function to generate a story
def generate_story(prompt, max_length=300, num_return_sequences=1):
    # Encode the input prompt
    input_ids = tokenizer.encode(prompt, return_tensors='pt')

    # Generate story using the model
    outputs = model.generate(
        input_ids,
        max_length=max_length,
        num_return_sequences=num_return_sequences,
        no_repeat_ngram_size=2,
        repetition_penalty=2.5,
        top_p=0.92,
        top_k=50,
        temperature=1.2,
        do_sample=True,
        early_stopping=True
    )

    # Decode the generated story
    stories = [tokenizer.decode(output, skip_special_tokens=True) for output in outputs]
    return stories

# Example prompt
prompt = "Once upon a time in a land far away"

# Generate the story
stories = generate_story(prompt)

# Print the generated story
for i, story in enumerate(stories):
    print(f"Story {i+1}:\n{story}\n")
```
